```python
import pandas as pd
```

```python
df=pd.read_csv('/content/student_performance.csv')
```

```python
print(df)
```

```python
df.describe()
```

```python
print(f"Total Students: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")
print(f"Department Count: {df['department'].nunique()}")
```

```python
print(df['department'].value_counts())
```

```python
print(f"Highest Math Score:{df['math_score'].max()}")
print(f"Lowest Math Score:{df['math_score'].min()}")
print(f"Average Math Score:{df['math_score'].mean().round(2)}")
print(f"Average Programming Score:{df['programming_score'].mean().round(2)}")
print(f"Average Attendance Percentage:{df['attendance_percentage'].mean().round(2)}%")
```

```python
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']
gender_scores=df.groupby('gender')['total_score'].mean().round(2)
print(gender_scores)
```

```python
top5=df.sort_values('total_score',ascending=False).head(5)
for index,row in top5.iterrows():
  print(f"  {row['name']:<20}|{row['department']:<18}|Score:{row['total_score']}")
```

```python
low_att=df[df['attendance_percentage']>75][['name','department','attendance_percentage']]
if len(low_att)>0:
  print(low_att.to_string(index=False))
else:
 print("All students have attendance above 75%")
```
